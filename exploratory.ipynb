# Data Pre-processing

## Imports

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

## Loading Data

```python
folder_path = "./data/"
```

```python
file1_name = "crypto_data_updated_13_november.csv"
file2_name = "crypto_data_updated_27_august.csv"
```

## Exploring Parameters

```python
df1 = pd.read_csv(folder_path+file1_name)
```

```python
df1.head()
```

```python
df1["Date"] = pd.to_datetime(df1["Date"])
```

```python
df1["Date"] = df1['Date'].dt.tz_localize(None)
```

```python
df1.head()
```

```python
df1.info()
```

```python
df2 = pd.read_csv(folder_path+file2_name)
```

```python
df2["Date"] = pd.to_datetime(df2["Date"]);
```

```python
df2.head()
```

```python
df2.info()
```

```python
df1.isna().sum()
```

```python
df2.isna().sum()
```

```python
df1 = df1.set_index("Date")
df2 = df2.set_index("Date")
```

```python
columns = ["Close (BTC)","Close (ETH)","Close (USDT)","Close (BNB)"]
df1_ = df1[columns]
df1_.plot(figsize=(20,8));
```

```python
columns = ["Adj Close (BTC)","Adj Close (ETH)","Adj Close (USDT)","Adj Close (BNB)"]
df2_ = df2[columns]
df2_.plot(figsize=(20,8));
```
